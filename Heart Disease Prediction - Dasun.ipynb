```python
# A function for displaying categorical distribution - Generated by ChatGPT

# Styling:

cust_palt = [
    '#111d5e', '#c70039', '#f37121', '#ffbd69', '#ffc93c'
]

plt.style.use('ggplot')



def ctg_dist(df, cols, hue=None,rows=3, columns=3):
    
    
    fig, axes = plt.subplots(rows, columns, figsize=(16, 12))
    axes = axes.flatten()

    for i, j in zip(df[cols].columns, axes):
        sns.countplot(x=i,
                    data=df,
                    palette=cust_palt,
                    hue=hue,
                    ax=j,
                    order=df[i].value_counts().index)
        j.tick_params(labelrotation=10)
        
        total = float(len(df[i]))
        
        j.set_title(f'{str(i).capitalize()} Distribution')
        
        
        for p in j.patches:
            height = p.get_height()
            j.text(p.get_x() + p.get_width() / 2.,
                    height + 2,
                    '{:1.2f}%'.format((height / total) * 100),
                    ha='center')
        
        plt.tight_layout()
```

**VIF Test**

```python
from statsmodels.tools.tools import add_constant  
from statsmodels.stats.outliers_influence import variance_inflation_factor


def calc_VIF(x):
    vif= pd.DataFrame()
    vif['variables']=x.columns
    vif["VIF"]=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]

    return(vif)


X = add_constant(df)  
ds=pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])],index=X.columns)  
print(ds)
```

There is no significant Multicollinearity between feature Variables. Therefore univariate feature selections will be considered for feature selection

**ANOVA**

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif


fs = SelectKBest(score_func=f_classif, k=10)
X_selected = fs.fit_transform(X, Y)
print(X_selected.shape)
```

```python

feature_names = df.columns.tolist()
feature_names.remove('condition')
print(feature_names)
```

```python
#To determine the selected feature Variables from ANOVA test

selected_features = fs.get_support(indices=True)
selected_feature_names = [feature_names[i] for i in selected_features]
print("Selected Features:", selected_feature_names)
```

resting_blood_pressure, cholesterol,fasting_blood_sugar have been dropped during anova feature selection.

# Modeling

**Data Splitting**

```python
df.head()
```

```python
Y.value_counts()
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split


X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size=0.3 , random_state=42)
```

**Data Normalization**

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

**Model Training**

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [100, 10, 1.0, 0.1, 0.01,0.001,0.0001],
    'solver': ['newton-cg', 'lbfgs', 'liblinear','saga'],
    'max_iter': [100, 200, 500, 1000,1500,2000]
}

lg = LogisticRegression()
```

Hyperparameter tuning with GridsearchCV

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(lg, param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, Y_train)
```

```python
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

print(best_params)
print(best_estimator)
```

**Model training based on hyperparameter tuning**

```python
lg = LogisticRegression(C=0.01, solver='newton-cg')
lg.fit(X_train , Y_train)
```

```python
# make predictions on test set
y_pred=lg.predict(X_test)

print('Training set score: {:.4f}'.format(lg.score(X_train, Y_train)))

print('Test set score: {:.4f}'.format(lg.score(X_test, Y_test)))
```

```python
# MSE & RMSE 

from sklearn.metrics import mean_absolute_error , mean_absolute_percentage_error , mean_squared_error , accuracy_score
import math


mse_lg =mean_squared_error(Y_test, y_pred)
print('Mean Squared Error : '+ str(mse_lg))
rmse_lg = math.sqrt(mean_squared_error(Y_test, y_pred))
print('Root Mean Squared Error : '+ str(rmse_lg))
```

```python
accuracy_lg = accuracy_score(Y_test, y_pred)
print(accuracy_lg)
```

```python
# calculating and plotting the confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
from mlxtend.plotting import plot_confusion_matrix



cm1_lg = confusion_matrix(Y_test,y_pred)
plot_confusion_matrix(conf_mat=cm1_lg,show_absolute=True,
                                show_normed=True,
                                colorbar=True)
plt.show()
```

```python
#Classification Report

matrix = classification_report(Y_test,y_pred )
print(matrix)
```
